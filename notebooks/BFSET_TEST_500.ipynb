## 2. Verify Dataset Structure (400 train / 100 val)

```python
TEST_PATH = "/content/drive/MyDrive/BFSET_TEST"

print("Existe BFSET_TEST :", os.path.exists(TEST_PATH))
print("Train images :", len(os.listdir(TEST_PATH + "/images/train")))
print("Val images   :", len(os.listdir(TEST_PATH + "/images/val")))
```

## 3. Training YOLOv8 on BFSET_TEST

```python
yaml_path = "/content/drive/MyDrive/BFSET_TEST/bfset_test.yaml"

yaml_content = """
path: /content/drive/MyDrive/BFSET_TEST
train: images/train
val: images/val
names:
  0: beard
"""

with open(yaml_path, "w") as f:
    f.write(yaml_content)

print("bfset_test.yaml créé :", yaml_path)
```

```python
DATA_CONFIG = "/content/drive/MyDrive/BFSET_TEST/bfset_test.yaml"
model = YOLO("yolov8n.pt")

results = model.train(
    data=DATA_CONFIG,
    imgsz=640,
    epochs=40,
    batch=8,
    device='cpu',
    project="/content/drive/MyDrive/BFSET_TEST/runs_yolo",
    name="yolov8n_test500",
)
```

## 4. Evaluation Metrics (Precision, Recall, F1, mAP)


```python
metrics = model.val(data=DATA_CONFIG)
metrics
```

```python
simple = metrics.results_dict
simple
```

## Summary of Results

- Precision: 0.976  
- Recall: 0.829  
- F1-score: 0.897  
- mAP50: 0.911  
- mAP50-95: 0.728  

These results validate the practical reliability of the BFSET annotations on a small-scale subset (500 images).
